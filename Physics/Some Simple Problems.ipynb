# The Schrödinger equation

We write down the classical equation of motion in terms of the total energy E of the system, namely as the sum of kinetic ($T$) and potential ($V$) energy:

$$
E = T+V = \frac{p^2}{2m} + V
$$

The quantum mechanical description is found by substituting $E$ with $i \hbar \frac{\partial}{\partial t}$ and $p$ with $-i \hbar \nabla$ so that we have:

$$
i h \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \nabla^2 \psi + V\psi
$$

Which is the **Schrödinger equation**.

This equation is separable into two parts if we substitute $\psi(\textbf{r},t)=\phi(t)u(\textbf{r})$ and then divide by $\psi(\textbf{r},t)$ then:

$$
\frac{i \hbar}{\phi(t)}\frac{d\phi}{dt} = -\frac{\hbar^2}{2mu(\textbf{r})}\nabla^2u(\textbf{r})+V(\textbf{r}) = E
$$

Since the left side does not depend on $\textbf{r}$ and the right side does not depend on $t$ then must both be equal to the same constant, $E$. Hence we have:

$$
\phi(t) = \phi_0e^{-iEt/h}
$$
and
$$
Eu(\textbf{r}) = -\frac{\hbar^2}{2m}\nabla^2 u(\textbf{r}) + V(\textbf{r})u(\textbf{r})
$$

The first governs the time evolution of the quantum mechanical system. 

The second is called the **time-independent Schrödinger equation**. We interpret $u^*(\textbf{r})u(\textbf{r})$ as the **probability density** of finding the particle at $\textbf{r}$. Integrating over some finite volumes gives the probability of finding the particle there. It can be written as an operator equation:

$$
Eu(\textbf{r}) = (-\frac{\hbar^2}{2m}\nabla^2  + V)u(\textbf{r}) = Hu(\textbf{r})
$$
from which we can identify the **Hamiltonian** operator, $H$, so called because of its similarity to to Hamiltonian form of classical mechanics.
$$
H = (-\frac{\hbar^2}{2m}\nabla^2  + V)
$$

It follows that **the possible energy values E must be eigenvalues of the Hamiltonian operator**. Finding them is a big part of theory behind quantum mechanics.

## The particle in a box

Apart from the free particle where $V=0$ everywhere, the simplest problem we can solve is the particle in a 1-d box. Since we are dealing with only one dimension $\textbf{r} \rightarrow x$ and we take $V(x)=0$ if $|x|<a$ (the _interior_ of the box) and $V(x)=\infty$ everywhere else.

Outside the region we have $V(x)=\infty$, so for finite E, the equation can only be satisfied  when $u(x)=0$. Easy.

In the interior of the box $V(x)=0$ so the Hamiltonian operator in one dimension simply becomes:

$$
H = -\frac{\hbar^2}{2m}\nabla^2 + 0 = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2}
$$

The possible energy levels are the eigenvalues of this operator. Trying a solution of the form:

$$
u(x) = Asin(\alpha x)+Bcos(\alpha x)
$$

$$
Hu = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2}[Asin(\alpha x)+Bcos(\alpha x)]=-\frac{\hbar^2}{2m}[-A\alpha^2 sin(\alpha x)-B\alpha^2cos(\alpha x)]
$$

This is an identity if and only if 
$$
\frac{\hbar^2\alpha^2}{2m}=E \quad or \quad \alpha=\sqrt{\frac{2mE}{\hbar^2}}
$$

Applying the additional boundary conditions that $u(-a)=u(a)=0$. This can only happen when either $A=0$ and $u(x)=Bcos(\frac{n\pi}{2a}x)$ where n is an odd integer or $B=0$ and $u(x)=Asin(\frac{n\pi}{a}x)$ where n is an even integer.

The energy values are therefore given by:

$$
E_n=\frac{\alpha^2\hbar^2}{2m}=\frac{n^2\pi^2\hbar^2}{8a^2m} \quad \forall n=1,2,3,... 
$$

Normalising to find the remaining constant we find:

$$
\int_{\infty}^{-\infty} u^*(x)u(x) dx = \int_{a}^{-a} u^2(x) dx = 1
$$

This yields $A=B=\frac{1}{\sqrt{a}}$.

Putting it all together and setting $a=1$ we can take a look at the results for different energy levels.

```julia
using Plots
```

```julia
u(n) = (x) -> iseven(n) ? sin(n*π*x/2) : cos(n*π*x/2)
P(u) = (x) -> u(x)^2
```

```julia
plot(-1:0.01:1, [P(u(n)) for n=1:6], fill=true, layout=6, leg=:none, title = permutedims("n=".*string.(collect(1:6))))
```
