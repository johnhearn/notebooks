### Plot the cumulative mean of samples from a power law (Pareto) distribution.

```julia
using Statistics, Plots, Distributions
```

```julia
# Calculate the cum mean over T samples
function sampler(T, sample)
    E = [sample()]
    for i in 1:T-1
        r = sample()
        push!(E, E[i] + (r - E[i])/i)
    end
    E
end

# Construct an ensemble of N running means over a T random samples from the given distribution.
function ensemble(N, T, dist)
    Es = []
    for k in 1:N
        push!(Es, sampler(T, () -> rand(dist)))
    end
    Es
end    
```

```julia
p1 = plot(ensemble(10, 10000, Pareto(1,1.5)), title="Pareto(1,3/2)")
p2 = plot(ensemble(10, 10000, Exponential()), title="Exponential(1)", )
plot(p1, p2, size=(900,400), alpha=0.5, lc=:gray, leg=:none, framestyle = :box, xaxis="Number of samples taken", yaxis="Running mean")
```

Note how the mean of samples from the Pareto distribution is characterised by low probability high impact samples, the spikes in the plot. For this reason it is one member of a family of distributions called *fat tailed* or *heavy-tailed* and the spikes are called *tail events*. The mean *never* converges even though it might seem to and you cannot predict when a new tail event might occur.

In comparison, a well behaved distribution like the Exponential has a mean that converges over time without spikes.
