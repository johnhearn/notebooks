Using a contraint solver to solve my friend David's son's ridiculously hard maths problem.

![The Problem](davids-fruit-problem.jpg)

```python
from constraint import *
```

```python
problem = Problem()
problem.addVariable("pear", range(1, 9))
problem.addVariable("kiwi", range(1, 9))
problem.addVariable("banana", range(1, 9))
problem.addVariable("apple", range(1, 9))
problem.addVariable("orange", range(1, 9))
problem.addVariable("cherry", range(1, 9))
problem.addVariable("strawberry", range(1, 9))
problem.addVariable("melon", range(1, 9))
problem.addVariable("watermelon", range(1, 9))

problem.addConstraint(
    lambda pear, kiwi, banana, apple: pear*(kiwi-banana) - apple == banana, 
    ("pear", "kiwi", "banana", "apple"))
problem.addConstraint(
    lambda orange, pear, cherry, kiwi: orange/pear + cherry - kiwi == kiwi, 
    ("orange", "pear", "cherry", "kiwi"))
problem.addConstraint(
    lambda strawberry, pear, apple, melon: strawberry*strawberry/pear + apple == melon, 
    ("strawberry", "pear", "apple", "melon"))
problem.addConstraint(
    lambda melon, kiwi, pear, banana, watermelon: (melon+kiwi)/pear + banana == watermelon, 
    ("melon", "kiwi", "pear", "banana", "watermelon"))

solutions = problem.getSolutions()
```
